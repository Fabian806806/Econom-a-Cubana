## El Verdadero Problema 

A modo de resumen , el dilema del ahorro en Cuba va más allá de la simple elección entre la moneda nacional o el dólar , la problemática principal recae en la imposibilidad del cubano promedio 

de contar con esos ingresos necesarios una vez cubiertas las necesidades básicas . Tristemente ahorrar deja de ser una decisión financiera y se convierte en un privilegio al alcance de muy pocos 

ciudadanos , situación que bajo las condiciones económicas actuales no parece tener solución a corto plazo independientemente de la moneda que se desee utilizar .
